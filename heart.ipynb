# <span style = "color:green"> clean data </span>

```python
df=df.drop_duplicates()
```

#### #duplicate records were removed 

```python
df=df.drop(columns=['restecg','thalach','exang','oldpeak','slope'])
```

#### #columns above were removed because they were not useful for the analysis 

```python
df['sex']=df['sex'].replace({0:'female',1:'male'})
```

#### #the value 0 was replaced with "female" and 1 with "male" to simplify the analysis

```python
df.info()
```

#### #the dataset was fully cleaned and now contains 302 rows and 9 columns 

# <span style = "color:green"> data analysis</span>

```python
df['sex'].value_counts().plot(kind='pie',figsize=(4,4),startangle=120,autopct='%1.1f%%',shadow=True)
plt.title('distribution sex',c='b')
plt.style.use('fivethirtyeight')
plt.show()
```

#### #the dataset contains twice as many males as females 

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df,x='sex',hue='target')
plt.title("effect sex on target",c='white')
plt.style.use('bmh')
plt.show()
```

#### #the data initially suggests that males have more cases than females , but this is misleading because the dataset contains twice as many males , after adjusting for this , females actually shows a higher proportion of cases 

```python
df.groupby('target')['chol'].mean()
```

```python
df.groupby('target')['age'].mean()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=df,x='target',hue='fbs')
plt.title('effect fbs on target',c='white')
plt.show()
```

#### #there is no clear relationship between fbs and the disease

```python
plt.figure(figsize=(9,3))
sns.countplot(data=df,x='target',hue='cp')
plt.title('effect cp on target')
plt.style.use('classic')
plt.show()
```

#### #both typical (0) and atypical (1) chest pain , whether heart-related (0) or not (2) , affect disease occurrence

```python
plt.figure(figsize=(9,3))
sns.countplot(data=df,x='target',hue='ca')
plt.title('effect ca on target')
plt.style.use('classic')
plt.show()
```

#### #no clear effect of major vessel damage on disease is observed , as target(1) is low at value 0

```python
pd.crosstab(df['ca'],df['target'],normalize='index')*100
```

```python
plt.figure(figsize=(9,3))
sns.countplot(data=df,x='target',hue='thal')
plt.title('effect thal on target')
plt.style.use('classic')
plt.show()
```

#### #blood flow defects (value 2) have a strong impact on disease , as shown in target (1)

```python
corr=df[['trestbps','age','chol']].corr()
```

```python
sns.heatmap(corr,annot=True,linewidth=3,cmap='coolwarm')
plt.title('effect age,chol,trestbps on target',c='r')
plt.show()
```

#### #there is no clear relationship between cholesterol(chol) , age , or resting blood pressure (trestbps) and the disease
